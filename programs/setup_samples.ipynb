```python
import glob
import random
import json
import os
import shutil
```

```python
%time all_files = glob.glob("../inputs/ActinobacteriaAnalysis_GBKs/*.cluster*.gbk")
len(all_files)
```

```python
all_files_not_in_contig_edge = []
for filepath in all_files:
    with open(filepath, "r") as handle:
        for line in handle:
            line = line.lstrip().rstrip()
            if line.startswith("/contig_edge="):
                if line.split("=")[1] == "\"False\"":
                    all_files_not_in_contig_edge.append(filepath)
                else:
                    print("{} is in contig edge, skipped.".format(filepath))
                break
```

```python
len(all_files_not_in_contig_edge)
```

```python
samples = {}

num = 10
while num <= 10000:
    samples[num] = []
    random.seed(123456789)
    for i in range(5):
        samples[num].append(random.sample(all_files_not_in_contig_edge, num))
    num = num * 10

if os.path.exists("../preprocessed/samples"):
    shutil.rmtree("../preprocessed/samples")
os.makedirs("../preprocessed/samples")

for n in samples:    
    for i, sample_list in enumerate(samples[n]):
        with open("../preprocessed/samples/sample-{:05d}-{:02d}.txt".format(n, i+1), "w") as handle:
            for filepath in sample_list:
                handle.write("{}\n".format(filepath))
```
